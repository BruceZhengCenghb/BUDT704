```python
# creat the list to contain all the cards
all_card = []

for n in range(2,11):

    for i in range(1,5):
        all_card.append(n)
    #print(all_card)
print("All cards we have are:", all_card)
#print(all_card[2] + all_card[4])


total_points = sum(all_card)
print("The total of all points is :", total_points)
print(len(all_card))
player1 = []
player2 = []

#function for cards distribution
def dist_cards(cards):  # "card" is the list input
    cards_avlb = cards   # cards available
    print(cards_avlb)
    n = len(cards_avlb)    # n count the cards left
    playerA = []
    playerB = []
    while n >= 1:
        x = random.randrange(1, (n+1))
        card_chose = cards_avlb[(x-1)]      #the card chosen from cards left
        if n%2 == 0:
            playerA = playerA + [card_chose]

        else:
            playerB = playerB + [card_chose]
        n = n - 1
        del(cards_avlb[(x-1)])
    print("The cards hold by two players are:", playerA, '\n', playerB)
    return playerA, playerB


def prove_fair(A, B):   # A, B is the lists of cards hold by AB
    print("The cards hold by two players are:", A, '\n', B)
    x = sum(A)  #sum of A
    y = sum(B)  #sum of B
    m = len(A)  #number of cards of A
    n = len(B)  #number of cards of B
    print('The sum of A and B cards are:', x, y)
    print('the number of cards hold by A B are:', m, n)
```

```python
def play_cards(A, B):   #input the two card list divided before
    hand1 = A       #player A
    hand2 = B       #player B
    r = 1           #round count
    out3 = []       #to contain the cards pulled out but Tie
    #print(hand2[0:2])

    while (len(hand1) > 0) and (len(hand2) > 0) and (r <= 25):
        out1 = hand1[0:2]  #cards give out by player 1
        out2 = hand2[0:2]
        total1 = out1[0] + out1[1]   #add two cards
        total2 = out2[0] + out2[1]
        r = r + 1

        print('player 1 shows ', out1, 'total is ', total2)
        print('player 2 shows ', out2, 'total is ', total2)
        print('This is the ', (r-1), ' round')

        del(hand1[0:2])   #delet the cards give out from the list
        del(hand2[0:2])
        

        if (total1 > total2):
            hand1 = hand1 + out1 + out2
            print('Player 1 wins the round and adds all four cards to their set of cards.')
            out3 = []
            
        elif (total1 < total2):
            hand2 = hand2 + out1 + out2
            print('Player 2 wins the round and adds all four cards to their set of cards.')
            out3 = []

        else:
            print('Tie! Next cards are pulled.')
            out3 = out3 + out1 + out2   # used to contain all the cards pulled out 

    final_score1 = sum(hand1)
    final_score2 = sum(hand2)
    if final_score1 > final_score2:
        print('player1 win with score', final_score1)
    else:
        print('player2 win with score', final_score2) 
```

```python
#part 2
# copy of the funcetions used

def play_cards_prove(A, B):   #input the two card list divided before
    hand1 = A       #player A
    hand2 = B       #player B
    r = 1           #round count
    out3 = []       #to contain the cards pulled out but Tie
    #print(hand2[0:2])


    while (len(hand1) > 0) and (len(hand2) > 0) and (r <= 25):
        out1 = hand1[0:2]  #cards give out by player 1
        out2 = hand2[0:2]
        total1 = out1[0] + out1[1]   #add two cards
        total2 = out2[0] + out2[1]
        r = r + 1
        start_total1 = sum(A[0:2])
        start_total2 = sum(B[0:2])

        # print('player 1 shows ', out1, 'total is ', total2)
        # print('player 2 shows ', out2, 'total is ', total2)
        # print('This is the ', (r-1), ' round')

        del(hand1[0:2])   #delet the cards give out from the list
        del(hand2[0:2])
        

        if (total1 > total2):
            hand1 = hand1 + out1 + out2
            # print('Player 1 wins the round and adds all four cards to their set of cards.')
            out3 = []
            
        elif (total1 < total2):
            hand2 = hand2 + out1 + out2
            # print('Player 2 wins the round and adds all four cards to their set of cards.')
            out3 = []

        else:
            #print('Tie! Next cards are pulled.')
            out3 = out3 + out1 + out2   # used to contain all the cards pulled out 

    final_score1 = sum(hand1)
    final_score2 = sum(hand2)
    if final_score1 > final_score2:
        #print('player1 win with score', final_score1)
        return final_score1, final_score2, start_total1, start_total2  #win final score, lose final score, win start score, lose start score
    else:
        #print('player2 win with score', final_score2) 
        #start_total = sum(B[0:2])
        return final_score2, final_score1, start_total2, start_total1
  
def dist_cards_prove(cards):  # "card" is the list input
    cards_avlb = cards   # cards available
    print(cards_avlb)
    n = len(cards_avlb)    # n count the cards left
    playerA = []
    playerB = []
    while n >= 1:
        x = random.randrange(1, (n+1))
        card_chose = cards_avlb[(x-1)]      #the card chosen from cards left
        if n%2 == 0:
            playerA = playerA + [card_chose]

        else:
            playerB = playerB + [card_chose]
        n = n - 1
        del(cards_avlb[(x-1)])
    #print("The cards hold by two players are:", playerA, '\n', playerB)
    return playerA, playerB


```
