The value without '-' is the main problem

```python
df3 = df2.copy()
```

```python

# df3[['asset class', 'tmp1']]= df3['Segment'].str.split(':', expand = True) #split them into 2 first
# df3.drop(columns=['Segment'],inplace= True)
```

```python
ser1 = df3['Segment'].str.contains(' - ')==False  # return ser of index do not have -
df3.loc[ser1, 'Segment'] = df3.loc[ser1, 'Segment'].str.replace('Global ','Global - ', regex=False)

df3.loc[ser1, 'Segment'] = df3.loc[ser1, 'Segment'].str.replace('U.S. ','U.S. - ', regex=False)
df3.loc[ser1, 'Segment'] = df3.loc[ser1, 'Segment'].str.replace('Energy ','Energy - ', regex=False)
df3.loc[ser1, 'Segment'] = df3.loc[ser1, 'Segment'].str.replace('Developed ','Developed - ', regex=False)
```

```python
# to check
ser2 = df3['Segment'].str.contains(' - ')==False 
ser2
```

```python
df3[['asset class', 'tmp']] = df3['Segment'].str.split(': ', expand = True)  #tmp first

df3[['market', 'segment']] = df3['tmp'].str.split(' - ', expand = True)
df3.drop(columns= ['tmp', 'Segment'], inplace= True)
df3.head()
```
